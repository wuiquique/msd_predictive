```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('../../msd_full.csv')

# df['duration'] = df['duration'] * 1000

df = df[df['year'] != 0]
df = df[df['popularity'] != 0]
df = df[df['tempo'] != 0]

df = df.dropna()
```

### Popularidad (?)

```python
plt.scatter(df['year'], df['popularity'])
plt.title('Dispersión de Popularidad por Año')
plt.xlabel('Año')
plt.ylabel('Popularidad')

coefficients = np.polyfit(df['year'], df['popularity'], 1)
poly_line = np.poly1d(coefficients)
plt.plot(df['year'], poly_line(df['year']), color='r')

slope = coefficients[0]

print(f"Pendiente: {slope}")

plt.grid(True)

plt.show()
```
