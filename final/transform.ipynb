```python
def divisor_tempo(tempo):
    if tempo <= 60:
        return 3
    elif tempo <= 65:
        return 2.5
    elif tempo <= 70:
        return 2
    elif tempo <= 80:
        return 1.5
    elif tempo >= 320:
        return 2
    elif tempo >= 300:
        return 1.6
    elif tempo >= 280:
        return 1.3
    elif tempo >= 260:
        return 1.2
    elif tempo >= 200 and tempo < 240:
        return 0.8
    elif tempo >= 160:
        return 2
    elif tempo >= 150:
        return 1.6
    elif tempo >= 140:
        return 1.3
    elif tempo >= 130:
        return 1.2
    elif tempo >= 100 and tempo <= 120:
        return 0.5
    elif tempo >= 90 and tempo < 100:
        return 0.65
    else:
        return 1

df["popularity"] = df.apply(
    lambda row: row["popularity"]
    if row["popularity"] > 0.85 and divisor_tempo(row["tempo"]) < 1
    else row["popularity"] / divisor_tempo(row["tempo"])
    if row["popularity"] / divisor_tempo(row["tempo"]) <= 0.95
    else row["popularity"] / (divisor_tempo(row["tempo"]) * 1.1)
    if row["popularity"] / (divisor_tempo(row["tempo"]) * 1.1) <= 0.95
    else row["popularity"] / (divisor_tempo(row["tempo"]) * 1.2)
    if row["popularity"] / (divisor_tempo(row["tempo"]) * 1.2) <= 0.95
    else row["popularity"] / (divisor_tempo(row["tempo"]) * 1.3)
    if row["popularity"] / (divisor_tempo(row["tempo"]) * 1.3) <= 0.95
    else row["popularity"] / (divisor_tempo(row["tempo"]) * 1.35)
    if row["popularity"] / (divisor_tempo(row["tempo"]) * 1.35) <= 0.95
    else row["popularity"] / (divisor_tempo(row["tempo"]) * 1.4)
    if row["popularity"] / (divisor_tempo(row["tempo"]) * 1.4) <= 1
    else row["popularity"],
    axis=1
)

df.describe()
```

```python
def divisor_loudness(loudness):
    if loudness <= -50:
        return 3
    elif loudness <= -30:
        return 2
    elif loudness <= -20:
        return 1.5
    elif loudness <= -10:
        return 1.2
    elif loudness > -10 and loudness <= 0:
        return 0.6
    elif loudness >= 1 and loudness <= 5:
        return 0.8
    elif loudness >= 10:
        return 1.3
    else:
        return 1

df["popularity"] = df.apply(
    lambda row: row["popularity"] / divisor_loudness(row["loudness"])
     if row["popularity"] / divisor_loudness(row["loudness"]) < 1
     else row["popularity"] / (divisor_loudness(row["loudness"]) * 1.1)
     if row["popularity"] / (divisor_loudness(row["loudness"]) * 1.1) < 1
     else row["popularity"] / (divisor_loudness(row["loudness"]) * 1.2)
     if row["popularity"] / (divisor_loudness(row["loudness"]) * 1.2) < 1
     else row["popularity"] / (divisor_loudness(row["loudness"]) * 1.3)
     if row["popularity"] / (divisor_loudness(row["loudness"]) * 1.3) < 1
     else row["popularity"] / (divisor_loudness(row["loudness"]) * 1.35)
     if row["popularity"] / (divisor_loudness(row["loudness"]) * 1.35) < 1
     else row["popularity"] / (divisor_loudness(row["loudness"]) * 1.4)
     if row["popularity"] / (divisor_loudness(row["loudness"]) * 1.4) < 1
     else row["popularity"]
     , axis=1
)

df.describe()
```

```python
def divisor_duration(duration):
    if duration <= 30:
        return 2.5
    elif duration <= 45:
        return 2.2
    elif duration <= 60:
        return 2
    elif duration <= 80:
        return 1.8
    elif duration <= 100:
        return 1.5
    elif duration <= 120:
        return 1.2
    elif duration >= 510:
        return 2.5
    elif duration >= 450:
        return 2.2
    elif duration >= 420:
        return 2
    elif duration >= 390:
        return 1.8
    elif duration >= 360:
        return 1.5
    elif duration >= 330:
        return 1.2
    elif duration >= 150 and duration <= 210:
        return 0.4
    else:
        return 1

df["popularity"] = df.apply(
    lambda row:
    row["popularity"] / divisor_duration(row["duration"])
    if row["popularity"] / divisor_duration(row["duration"]) <= 1
    else row["popularity"] / (divisor_duration(row["duration"]) * 1.1)
    if row["popularity"] / (divisor_duration(row["duration"]) * 1.1) <= 1
    else row["popularity"] / (divisor_duration(row["duration"]) * 1.2)
    if row["popularity"] / (divisor_duration(row["duration"]) * 1.2) <= 1
    else row["popularity"] / (divisor_duration(row["duration"]) * 1.3)
    if row["popularity"] / (divisor_duration(row["duration"]) * 1.3) <= 1
    else row["popularity"] / (divisor_duration(row["duration"]) * 1.35)
    if row["popularity"] / (divisor_duration(row["duration"]) * 1.35) <= 1
    else row["popularity"] / (divisor_duration(row["duration"]) * 1.4)
    if row["popularity"] / (divisor_duration(row["duration"]) * 1.4) <= 1
    else row["popularity"]
    , axis=1
)

df.describe()
```

```python
def divisor_key_mode(key, mode):
    if mode == 0: ## Menor
        if key == 9:
            return 0.4
        elif key == 2 or key == 4 or key == 0:
            return 0.6
        elif key == 5 or key == 1 or key == 3 or key == 8 or key == 10:
            return 1.4
        else:
            return 1
    elif mode == 1: ## Mayor
        if key == 0:
            return 0.5
        elif key == 2 or key == 7 or key == 9:
            return 0.7
        elif key == 11 or key == 1 or key == 3 or key == 6 or key == 8 or key == 10:
            return 1.2
        else:
            return 1

df["popularity"] = df.apply(
    lambda row:
    row["popularity"] / divisor_key_mode(row["key"], row["mode"])
    if row["popularity"] / divisor_key_mode(row["key"], row["mode"]) <= 1
    else row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.1)
    if row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.1) <= 1
    else row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.2)
    if row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.2) <= 1
    else row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.3)
    if row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.3) <= 1
    else row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.35)
    if row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.35) <= 1
    else row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.4)
    if row["popularity"] / (divisor_key_mode(row["key"], row["mode"]) * 1.4) <= 1
    else row["popularity"]
    , axis=1
)

df.describe()
```
